```python
# 1.1 
import os
os.chdir('C:/Users/maria.DESKTOP-OVDQUT6/Desktop/docs/Python/HW_for_Python/poems/poems')
print(os.getcwd())
os.remove('poems.txt')
skolko = 0
for root, dirs, files in os.walk('.'):
    for name in files:
        fullname = os.path.join(root, name)
        print(fullname)
        skolko = skolko + 1
# 1.2
print(skolko)
# 2
with open('poems.txt','w',encoding='utf-8') as f:
    for root, dirs, files in os.walk('.'):
        for text in files:
            if os.path.exists(text):
                with open(text,'r',encoding='utf-8') as name:
                    name=name.read()
                    f.write(name+ '\n')
            elif os.path.exists('.\котики\%s' %text):
                with open('.\котики\%s' %text,'r',encoding='utf-8') as name:
                    name=name.read()
                    f.write(name+ '\n')
            elif os.path.exists('.\сказки\%s' %text):
                with open('.\сказки\%s' %text,'r',encoding='utf-8') as name:
                    name=name.read()
                    f.write(name+ '\n')
with open ('poems.txt','r',encoding='utf-8') as q:
    print (q.read())
```
